```python
def pai_loop_over_obs_slow(obs_list, vars, enslon, enslat, ep_values, mean_pres, nens):
    incr = [[vars[i], np.zeros((ep_values[i].shape[1], ep_values[i].shape[2]))] for i in range(len(vars))]
    #ep_values = [ens_perturb[var].values for var in vars]
    for i, obs in enumerate(obs_list):
        hgc = PAI.compute_hloc(obs["obslat"], obs["obslon"], obs["hloc"], enslat, enslon)
        #print(hgc.shape)
        #print(enslat.shape)
        if obs["vloc"] < 10:
            vgc = PAI.compute_vloc(obs["obspres"] / 100.0, obs["vloc"], mean_pres / 100.0)
        else:
            vgc = 1
        kcol = [PAI.compute_kalman_gain(
            obs["anaensT"],
            obs["err"],
            ep_values[k],
            nens,
            hgc=hgc,
            vgc=vgc
        ) for k in range(len(vars))]
        #print(obs['obslat'])
        #print(PAI.compute_kalman_gain(
        #    obs["anaensT"],
        #    obs["err"],
        #    ep_values[0],
        #    nens,
        #    hgc=hgc,
        #    vgc=vgc
        #))
        #print(obs['anaensT'].shape)
        #print(hgc.shape)
        #print(vgc)
        big_K = 1.0 / (nens - 1) * PAI.K_column(ep_values[0], obs["anaensT"], obs["err"] ** (-2))
        #print(big_K.shape)
        #print(kcol[0].shape)
        print(incr[0][1].shape)
        #print(vgc*hgc*big_K)
        #print(ep_values[0].shape)
        for l in range(len(vars)):
            #print(kcol[0].shape)
            #print(incr[l][1].shape, PAI.compute_increment(kcol[l], obs["fgmeandep"]).shape)
            incr[l][1] += PAI.compute_increment(kcol[l], obs["fgmeandep"])
    return incr
```

```python
def explanatory_plot_split(exp_path, dir_time, ana_time, sonde_time, rep, rep_index, gc_plevel, gc_width, save_name, sat_var = 'REFL', RTPP = True, Title = 'Explanatory_Plot', surface_vloc = 0.2, Surface_Peak = False, vars = ['t', 'q'], obs_var = 'T', save = True):    
    label_size = 23
    obs_path_RAD = '{}/feedback/{}/ekfRAD.nc.{}'.format(exp_path, ana_time, ana_time)
    obs_path_TEMP = '{}/feedback/{}/ekfTEMP.nc.{}'.format(exp_path, dir_time, sonde_time)
    ana_path = '{}/{}/an_R19B07.{}.mean'.format(exp_path, ana_time, ana_time)
    increment_path = '{}/{}/an_R19B07.{}_inc'.format(exp_path, ana_time, ana_time)
    obs_sonde_RH = paiut.get_ekf(obs_path_TEMP, "RH", active = False)
    obs_sonde_T = paiut.get_ekf(obs_path_TEMP, "T", active = False)
    obs_sonde_RH.add_filter(filter=f"report={rep}")
    obs_sonde_T.add_filter(filter=f"report={rep}")
    obslon = obs_sonde_RH.obs(param="lon")[0]
    print(obs_sonde_RH.obs(param="lon"))
    obslat = obs_sonde_RH.obs(param="lat")[0]
    #Get region close to the radiosonde and restrict analysis and radiances to this region.
    rot_lons, rot_lats = oi.location_to_rotated_pole([obslon, obslon], [obslat, obslat])

    infl_reg, obs_reg = paiut.get_regions(*rot_lons, *rot_lats, hloc=25.0)

    infl_reg_latlon = np.array(oi.rotated_pole_to_location(infl_reg[0:2], infl_reg[2:4])).flatten()#[list(array) for array in rotated_pole_to_location(infl_reg[0:2], infl_reg[2:4])]

    print("get obs batch")
    obs_batch = paiut.get_obs_batch(obs_path_RAD, *infl_reg_latlon, obsvar=sat_var, new_loc = False, RAWBT_choice = 0)#, RTPP_correction=RTPP)
    print(len(obs_batch))

    print("read data")
    var = 't'
    buffer = 0.04
    mean = paiut.read_grib(ana_path, vars).isel(time=0) 
    inc = paiut.read_grib(increment_path, vars).isel(time=0) 
    #ens = paiut.read_grib_mf(ana_path, ana_time, vars)
    #ens2 = loc.find_analysis_in_area(obslon - buffer, obslon + buffer, obslat - buffer, obslat + buffer, ens.clon, ens.clat, ens)
    mean2 = loc.find_analysis_in_area(obslon - buffer, obslon + buffer, obslat - buffer, obslat + buffer, mean.clon, mean.clat, mean)
    inc2 = loc.find_analysis_in_area(obslon - buffer, obslon + buffer, obslat - buffer, obslat + buffer, inc.clon, inc.clat, inc)
    fc = mean2 - inc2
    #ensperturb = ens2 - mean2
    if RTPP == True:
        ensperturb = xr.open_dataset('/jetfs/home/a12233665/pai-munich-vienna/Analysis_Forecast_Pert_Files/tilde_X_a_' + ana_time + '_' + str(rep_index))
    else:
        ensperturb = xr.open_dataset('/jetfs/home/a12233665/pai-munich-vienna/Analysis_Forecast_Pert_Files/X_a_' + ana_time + '_' + str(rep_index))
    
    #Extract values now so that we don't need to every time we call PAI.
    pressure_levels = mean2.pres.values
    lats = ensperturb.clat.values
    lons = ensperturb.clon.values
    perturbvars = [ensperturb[var].values for var in vars]
    perturbvars.append(ensperturb.pres.values)
    #Get actual increment (approximated by PAI) attributed to radiances in this area (for error calculations later).
    print('get actual increment t')
    incr_REAL = paiogc.pai_loop_over_obs_slow(
            obs_batch,
            ['t', 'q', 'pres'],
            lons,
            lats,
            perturbvars,
            pressure_levels,
            40,
        )
    incr_REAL_T = incr_REAL[0][1]
    incr_REAL_Q = incr_REAL[1][1]
    incr_REAL_P = incr_REAL[2][1]
    #Get increment for specific Gaspari-Cohn
    for obs in obs_batch:
        obs['vloc'] = gc_width
        obs['obspres'] = gc_plevel
    #Calculate PAI
    #print('PAI t')
    if Surface_Peak:
        incr = paiogc.pai_loop_over_obs_slow_surface_peak(
            obs_batch,
            ['t', 'q', 'pres'],
            lons,
            lats,
            perturbvars,
            pressure_levels,
            40,
        )
    else:
        incr = paiogc.pai_loop_over_obs_slow(
            obs_batch,
            ['t', 'q', 'pres'],
            lons,
            lats,
            perturbvars,
            pressure_levels,
            40,
        )
    incr_T = incr[0][1]
    incr_Q = incr[1][1]
    #print(incr_Q, mean2.q.values)
    incr_P = incr[2][1]
    #Get new analysis field with PAI increment.
    t_no_obs = mean2.t.values - incr_REAL_T
    q_no_obs = mean2.q.values - incr_REAL_Q
    p_no_obs = mean2.pres.values - incr_REAL_P
    #Interpolate
    start_time = time.time()
    print('start interpolating')
    mean_t_no_obs = LinearNDInterpolator(list(zip(lats, lons)), t_no_obs.T)(obslat, obslon)
    mean_q_no_obs = LinearNDInterpolator(list(zip(lats, lons)), q_no_obs.T)(obslat, obslon)
    mean_p_no_obs = LinearNDInterpolator(list(zip(lats, lons)), p_no_obs.T)(obslat, obslon)
    print(mean_t_no_obs.shape)

    print('start interpolating')
    incr_t_obsloc = LinearNDInterpolator(list(zip(lats, lons)), incr_T.T)(obslat, obslon)
    incr_q_obsloc = LinearNDInterpolator(list(zip(lats, lons)), incr_Q.T)(obslat, obslon)
    print('Finished second interpolation')
    incr_REAL_t_obsloc = LinearNDInterpolator(list(zip(lats, lons)), incr_REAL_T.T)(obslat, obslon)
    incr_REAL_q_obsloc = LinearNDInterpolator(list(zip(lats, lons)), incr_REAL_Q.T)(obslat, obslon)

    ana_obsloc_t = LinearNDInterpolator(list(zip(lats, lons)), mean2.t.values.T)(obslat, obslon)
    ana_obsloc_q = LinearNDInterpolator(list(zip(lats, lons)), mean2.q.values.T)(obslat, obslon)
    fc_obsloc_t = LinearNDInterpolator(list(zip(lats, lons)), fc.t.values.T)(obslat, obslon)
    fc_obsloc_q = LinearNDInterpolator(list(zip(lats, lons)), fc.q.values.T)(obslat, obslon)
    print(ana_obsloc_q, incr_q_obsloc)
    interp_plevs = LinearNDInterpolator(list(zip(lats, lons)), pressure_levels.T)(obslat, obslon)
    #Interpolate to radiosonde pressure levels.

    incr_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), interp_plevs, incr_t_obsloc)
    incr_t_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, incr_t_obsloc)
    incr_Q_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, incr_q_obsloc)

    incr_REAL_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), interp_plevs, incr_REAL_t_obsloc)
    incr_REAL_Q_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, incr_REAL_q_obsloc)

    ana_t_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, ana_obsloc_t)
    #print('t ana', ana_t_obs_plevs)
    ana_q_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, ana_obsloc_q)
    #print('q ana', ana_q_obs_plevs)
    ana_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), interp_plevs, ana_obsloc_t)

    fc_t_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, fc_obsloc_t)
    #print('t ana', ana_t_obs_plevs)
    fc_q_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), interp_plevs, fc_obsloc_q)
    #print('q ana', ana_q_obs_plevs)
    fc_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), interp_plevs, fc_obsloc_t)

    no_obs_t_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), mean_p_no_obs, mean_t_no_obs)
    #print('t no obs', no_obs_t_obs_plevs)
    no_obs_q_obs_plevs = np.interp(np.flip(obs_sonde_RH.obs(param = 'plevel')), mean_p_no_obs, mean_q_no_obs)
    #print('q no obs', no_obs_q_obs_plevs)
    no_obs_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), mean_p_no_obs, mean_t_no_obs)

    with_loc_t_obs_plevs = no_obs_t_obs_plevs + incr_t_obs_plevs
    #print('t loc', with_loc_t_obs_plevs - 273.15)
    with_loc_q_obs_plevs = no_obs_q_obs_plevs + incr_Q_obs_plevs
    #print('q loc', with_loc_q_obs_plevs)
    with_loc_T_obs_plevs = no_obs_T_obs_plevs + incr_T_obs_plevs

    no_obs_RH_obs_plevs = np.array([relative_humidity_from_specific_humidity(np.flip(obs_sonde_RH.obs(param = 'plevel'))[i]/100 * units.hPa, 
                                                                            (no_obs_t_obs_plevs[i] - 273.15) * units.degC, 
                                                                            no_obs_q_obs_plevs[i] * units('kg/kg')).to('percent').magnitude / 100 for i in range(len(no_obs_t_obs_plevs))])
    print(no_obs_RH_obs_plevs)
    ana_RH_obs_plevs = np.array([relative_humidity_from_specific_humidity(np.flip(obs_sonde_RH.obs(param = 'plevel'))[i]/100 * units.hPa,
                                                                        (ana_t_obs_plevs[i] - 273.15) * units.degC,
                                                                        ana_q_obs_plevs[i] * units('kg/kg')).to('percent').magnitude / 100 for i in range(len(ana_t_obs_plevs))])
    print(ana_RH_obs_plevs)
    fc_RH_obs_plevs = np.array([relative_humidity_from_specific_humidity(np.flip(obs_sonde_RH.obs(param = 'plevel'))[i]/100 * units.hPa,
                                                                        (fc_t_obs_plevs[i] - 273.15) * units.degC,
                                                                        fc_q_obs_plevs[i] * units('kg/kg')).to('percent').magnitude / 100 for i in range(len(ana_t_obs_plevs))])
    with_loc_RH_obs_plevs = np.array([relative_humidity_from_specific_humidity(np.flip(obs_sonde_RH.obs(param = 'plevel'))[i]/100 * units.hPa,
                                                                        (with_loc_t_obs_plevs[i] - 273.15) * units.degC,
                                                                        with_loc_q_obs_plevs[i]* units('kg/kg')).to('percent').magnitude / 100 for i in range(len(with_loc_t_obs_plevs))])
    print(with_loc_RH_obs_plevs)        
    no_obs_T_obs_plevs = np.interp(np.flip(obs_sonde_T.obs(param = 'plevel')), mean_p_no_obs, mean_t_no_obs)

    plt.rcParams.update({
        "text.usetex": False,
    })
    fig, ax = plt.subplots(figsize=(5, 9))
    if obs_var == 'T':
        ax.plot(incr_T_obs_plevs, np.flip(obs_sonde_T.obs(param = 'plevel'))/100, label='PAI $^{VIS \\to T}_\\rho$', color = 'r') #label='PAI with Retr. loc', color = 'r')
        ax.plot(incr_REAL_T_obs_plevs, np.flip(obs_sonde_T.obs(param = 'plevel'))/100, label = 'PAI $^{VIS \\to T}$', color = 'b')#label='Real PAI Vector', color = 'b')
        #ax.plot(ana_T_obs_plevs - fc_T_obs_plevs, np.flip(obs_sonde_T.obs(param = 'plevel'))/100, label='Full Increment', color = 'g')
        scale = max(max(incr_T_obs_plevs), max(incr_REAL_T_obs_plevs))
        sonde_error_no_obs = np.flip(obs_sonde_T.obs()) - no_obs_T_obs_plevs
        sonde_error_with_loc = np.flip(obs_sonde_T.obs()) - with_loc_T_obs_plevs
        sonde_error_real_obs = np.flip(obs_sonde_T.obs()) - ana_T_obs_plevs
        obs_sonde = obs_sonde_T
    else:
        ax.plot(with_loc_RH_obs_plevs - no_obs_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='PAI $^{VIS \\to r}_\\rho$', color = 'brown')
        ax.plot(ana_RH_obs_plevs - no_obs_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='PAI $^{VIS \\to r}$', color = 'g')
        #ax.plot(ana_RH_obs_plevs - fc_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='Full Increment', color = 'g')
        scale = max(max(ana_RH_obs_plevs - no_obs_RH_obs_plevs), max(with_loc_RH_obs_plevs - no_obs_RH_obs_plevs))
        sonde_error_no_obs = np.flip(obs_sonde_RH.obs()) - no_obs_RH_obs_plevs
        sonde_error_with_loc = np.flip(obs_sonde_RH.obs()) - with_loc_RH_obs_plevs
        sonde_error_real_obs = np.flip(obs_sonde_RH.obs()) - ana_RH_obs_plevs
        obs_sonde = obs_sonde_RH
    if Surface_Peak:
        surface_vgc = PAI.compute_vloc(1000, surface_vloc, np.flip(obs_sonde.obs(param = 'plevel')) / 100.0)
        if obs["vloc"] < 10:
            surface_influence = PAI.compute_vloc(1000, surface_vloc, np.array([gc_plevel]) / 100.0)
            gc = (PAI.compute_vloc(gc_plevel / 100.0, gc_width, np.flip(obs_sonde.obs(param = 'plevel')) / 100.0) + surface_vgc)/(1+surface_influence)
        else:
            gc = np.ones(len(obs_sonde.obs(param = 'plevel')))
    else:
        gc =PAI.compute_vloc(gc_plevel/100, gc_width, np.flip(obs_sonde.obs(param = 'plevel'))/100)
    ekf_RAD = paiut.get_ekf(obs_path_RAD, sat_var)
    gc_original = PAI.compute_vloc(ekf_RAD.obs(param='plevel')[0]/100, ekf_RAD.obs(param='v_loc')[0], np.flip(obs_sonde.obs(param = 'plevel'))/100)
    ax.plot(gc*scale, np.flip(obs_sonde.obs(param = 'plevel'))/100, linestyle = 'dotted', label="New loc $\\rho$, \n scaled", color = 'y')
    ax.plot(gc_original*scale, np.flip(obs_sonde.obs(param = 'plevel'))/100, linestyle = 'dotted', label="Old loc, \n scaled", color = 'purple')
    #ax.legend(loc = 'upper left', fontsize = label_size - 4) #, bbox_to_anchor=(0.55, 0.9),
    fig.legend(ncol = 2, bbox_to_anchor=(0.95, -0.03), fontsize = label_size - 1)
    ax.set_title('PAI original and localised \n Increment, ' + obs_var, fontsize = label_size + 2)
    if obs_var == 'T':
        ax.set_xlabel('T Increment, K', fontsize = label_size)
    else:
        ax.set_xlabel('RH Increment, Fraction', fontsize = label_size)
    if obs_var == 'T':
        ax.set_ylabel('Pressure Level (hPa)', fontsize = label_size)
    ax.invert_yaxis()
    ax.set_ylim([1000, 100])
    ax.axvline(x = 0, color = 'black', alpha = 0.5)
    ax.tick_params(axis='both', which='major', labelsize=15)
    if save:
        plt.savefig(save_name + '_Increments.pdf', bbox_inches='tight')

    fig, ax = plt.subplots(figsize=(5, 9))
    #axs[1].plot(sonde_error_no_obs, np.flip(obs_sonde.obs(param = 'plevel'))/100, label='Radiosonde - analysis without obs, ' + obs_var, color = 'g')
    #axs[1].plot(sonde_error_with_loc, np.flip(obs_sonde.obs(param = 'plevel'))/100, label='Radiosonde - analysis with loc, ' + obs_var, color = 'black')
    ax.plot(np.abs(sonde_error_no_obs) - np.abs(sonde_error_with_loc), np.flip(obs_sonde.obs(param = 'plevel'))/100, label='|Error no obs| - |Error new loc|, ' + obs_var , color = 'orange')
    ax.plot(np.abs(sonde_error_no_obs) - np.abs(sonde_error_real_obs), np.flip(obs_sonde.obs(param = 'plevel'))/100, label='|Error no obs| - |Error old loc|, ' + obs_var , color = 'cyan')
    ax.legend(loc = 'upper right', fontsize = label_size - 4)
    ax.set_title('Observation Benefit, ' + obs_var, fontsize = label_size + 2)
    if obs_var == 'T':
        ax.set_xlabel('T benefit, K', fontsize = label_size)
    else:
        ax.set_xlabel('RH benefit, Fraction', fontsize = label_size)
    ax.invert_yaxis()
    ax.set_ylim([1000, 100])
    if obs_var == 'T':
        ax.set_ylabel('Pressure Level (hPa)', fontsize = label_size)
    ax.axvline(x = 0, color = 'black', alpha = 0.5)
    ax.tick_params(axis='both', which='major', labelsize=label_size - 2)
    if save:
        plt.savefig(save_name + '_Benefit.pdf', bbox_inches='tight')

    fig, ax = plt.subplots(figsize=(5, 9))
    ax.plot(np.flip(obs_sonde_RH.obs()), np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='Radiosonde RH', color = 'b')
    ax.plot(no_obs_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='Model RH with Satellites \n Removed by PAI', color = 'r')
    ax.plot(ana_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='Analysis RH with Satellites', color = 'g')
    ax.plot(fc_RH_obs_plevs, np.flip(obs_sonde_RH.obs(param = 'plevel'))/100, label='Background RH', color = 'orange')    
    #ax.plot(np.flip(obs_sonde_T.obs()) - ana_T_obs_plevs, np.flip(obs_sonde_T.obs(param = 'plevel'))/100, label='Radiosonde T - Analysis T', color = 'b')
    #ax.plot(incr_T_obs_plevs, np.flip(obs_sonde_T.obs(param = 'plevel'))/100, label='PAI with Retr. loc', color = 'r')
    #ax.legend(loc = 'upper right', fontsize = label_size - 4)
    fig.legend(bbox_to_anchor=(0.95, -0.03), fontsize = label_size - 1)
    ax.set_title('RH plot', fontsize = label_size + 2)
    ax.set_xlabel('Relative Humidity, Fraction', fontsize = label_size)
    ax.set_ylabel('Pressure Level (hPa)', fontsize = label_size)
    ax.invert_yaxis()
    ax.set_ylim([1000, 100])
    ax.tick_params(axis='both', which='major', labelsize=label_size - 2)
    if save:
        plt.savefig(save_name + '_Profiles.pdf', bbox_inches='tight')
```

```python
def rotated_pole_to_location(rlons, rlats):
    #
    rotated_pole = ccrs.RotatedPole(pole_latitude=40, pole_longitude=-170)
    rotated_coords = [
        ccrs.PlateCarree().transform_point(rlon, rlat, rotated_pole)
        for (rlon, rlat) in zip(rlons, rlats)
    ]

    lons, lats = zip(*rotated_coords)

    return np.array(lons), np.array(lats)
```

```python
def read_grib(path, vars, grid_path="/jetfs/home/a12233665/pai-munich-vienna/assets/icon_grid_0047_R19B07_L.nc"):

    #Reads a grib file and assigns it standard lat/lon coordinates based on the grid file.
    data = read(path)#.rename(values="cell")
    data = data[[*vars, 'pres']]
    grid = xr.open_dataset(grid_path)

    data = data.assign_coords(clat=("cell2", grid.clat.values * 180 / np.pi))
    data = data.assign_coords(clon=("cell2", grid.clon.values * 180 / np.pi))
    data = data.rename_dims({"cell2": "cell"})

    return data
```
